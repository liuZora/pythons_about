# 主题：Pandas的数据统计函数
# 一、汇总类统计

```python
import pandas as pd
file_path = r"D:\JupyterProject\study-pandas\file\datas\weather_20230115134249.csv"
df = pd.read_csv(file_path)
print(df.head())
```

```python
#替换温度后面的
df.loc[:, "气温(度)"] = df["气温(度)"].str.replace("℃", "").astype('float32')
```

- df.describe():一下子提取所有数字列统计结果
- df[column_name].mean()/max()/min():平均值/最大值/最小值

```python
print(df.describe())
```

```python
print(f"气温平均值：{df['气温(度)'].mean()}, 气温最大值：{df['气温(度)'].max()},气温最小值:{df['气温(度)'].min()}")
```

以上都是数值类型的统计，那非数值类型呢？
- 唯一性去重unique()：一般不用于数值列，而是枚举、分类列。作用是查看某一列的全部取值。
- 按值计数value_counts():查看某一列每个值分别出现多少次。

```python
df['城市'].unique()
```

```python
df["气温(度)"].unique()
```

```python
print(df["气温(度)"].value_counts())
print(df['城市'].value_counts())
```
