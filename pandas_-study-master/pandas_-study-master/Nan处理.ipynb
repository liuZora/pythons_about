# Pandas对缺失值的处理

```python
import pandas as pd
stu_df = pd.read_excel(r"D:\JupyterProject\study-pandas\file\datas\student.xlsx", skiprows = 2)
print(stu_df)
```

```python
print(stu_df.isnull())
```

```python
print(stu_df["分数"].isnull())
print(stu_df["分数"].notnull())
```

```python
stu_df.dropna(axis = "columns", how = "all", inplace = True) #按照列 如果某列全部都是空值 那就删除这一列
print(stu_df)
```

```python
stu_df.dropna(axis = "rows", how = "all", inplace = True) #按照行 如果某列全部都是空值 那就删除这一列
print(stu_df)
```

```python
#将分数列为空的值填充为0分
#方法一
stu_df.fillna({"分数": 0}, inplace = True)
print(stu_df)
```

```python
#方法二
stu_df.loc[:, '分数'] = stu_df['分数'].fillna(0)
print(stu_df)
```

```python
#姓名缺失值填充
stu_df.loc[:, "姓名"] = stu_df["姓名"].fillna(method = "ffill") #ffill forward fill
print(stu_df)
```

```python
#保存处理好的形式
stu_df.to_excel("./clean1.xlsx", index = False)
```
