# 二、Pandas数据结构

## 2.1 DataFrame和Series

**DataFrame**是一个二维数据，整个表格，多行(对应索引列df.index)多列（对应df.columns）。

**Series**是一个一维数据，一行或一列。由一组数据（可以是不同类型数据）和一组与之相关的数据标签（索引）组成。

## 2.2 Series创建方式

```python
s1 = pd.Series([1, 'a', 5.2, 7])
#左侧为索引(默认为数据序列) 右侧为数据
print(s1, s1.index)
```

1. 创建一个具有标签索引的Series

```python
s2 = pd.Series([1, 'a', 5.2, 7], index = ['d', 'd', 'a', 'c'])
print(s2, s2.index)
```

2. 使用Python字典创建Series

```python
sdata = {'Ohio': 35000000, 'Ohio': 72000, 'Oregon': 16000, 'Utah': 5000}
s2 = pd.Series(sdata)
s2
```

## 2.3 根据标签查询数据
由以下例子可知， 索引是可以重复的。在根据索引查询时，会找出所有与之对应的值

```python
s2 = pd.Series([1, 'a', 5.2, 7], index = ['d', 'd', 'a', 'c'])
print(s2)
print(s2['d'])
```

根据多个索引查询有以下两点。
- 传参数时，要通过列表把要查询的索引包括在一起
- 查询结果是Series类型数据

```python
print(s2[['d', 'a',]])
print(type(s2[['a', 'd']]))
```

## 2.4 DataFrame
**DataFrame有以下三点**
- 每列可以是不同数据类型
- 既有行索引index，又有列索引columns
- 可以被看成是由Series组成的字典

### 2.4.1 创建方式

```python
# 列名为键 列表为值
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
    'year': [2000, 2001, 2002, 2001, 2002],
    'pop': [1.5, 1.7, 3.6, 2.4, 2.9]
}
df = pd.DataFrame(data)
print(df,df.index, df.columns, df.dtypes, sep = "\n")
```

### 2.4.2 从DataFrame种查询出Series
- 查询一列结果是一个pd.Series
- 查询多列，结果是一个pd.DataFrame

```python
print(df['year'], type(df['year']))
```

```python
print(df[['year', 'pop']], type(df[['year', 'pop']]))
```

如果查询行呢？使用下面两个方法！
- df.loc[1], 结果是一个Series
- df.loc[1:3], 结果是一个DataFrame,查询是**闭区间**

```python
print(df.loc[1])
```

```python
print(df.loc[1:3])
```
