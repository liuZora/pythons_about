# Pandas数据排序
# 一、Series的排序
**Series的排序**
Series.sort_values(ascending = True, inplace = False)
**DataFrame**
DataFrame.sort_values(by, ascending = True, inplace = False)
- by:字符串或者List[string]

```python
import pandas as pd
file_path = r"D:\JupyterProject\study-pandas\file\datas\weather_20230115134249.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
sorted_series = df["相对湿度(%)"].sort_values()
```

```python
print(sorted_series)
```

```python
df.sort_values(by = ["相对湿度(%)"])
print(df)
```

```python
df.sort_values(by = ["相对湿度(%)"], inplace = True)
print(df)
```

# 二、Pandas字符串处理
## 2.1 Pandas字符串处理
1. 使用方法：先获取Series的str属性，然后在属性上调用函数
2. 只能用于字符串
3. DataFrame没有
4. 链接：https://pandas.pydata.org/pandas-docs/stable/reference/series.html#string-handling
## 2.2 应用
### 2.2.1 获取Series的str属性，然后使用各种字符串处理函数
- 字符串替换replace:返回一个替换后的Series
- startswith
- slice方法
- 正则表达式

```python

import pandas as pd
file_path = r"D:\JupyterProject\study-pandas\file\datas\weather_20230115134249.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
#字符串替换
df["气温(度)"].str.replace("℃", "")
```

```python
condition = df["日期"].str.startswith("2015-03")
condition
```

```python
df[condition].head()
```

```python
#slice 比如获取日期的月份
df["日期"].str.slice(5, 7)
```

```python
def get_col(x):
    year, month, day = x["日期"].split('-')
    return f"{year}年{month}月{day}日"
df.loc[:, "中文日期"] = df.apply(get_col, axis = 1)
print(df)
```

# 三、axis参数如何理解？
- axis = 0:当操作是单行操作 那么指的是操作某一行；如果操作是聚合操作，那么指的是跨行操作。axis为1时同理。

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    np.arange(12).reshape(3, 4), 
    columns = ['A', 'B', 'C', 'D']
)
df
```

```python
#删除A这一列 单列操作 所以axis=1
new_df = df.drop('A', axis = 1)
print(new_df)
```

```python
#mean聚合操作 计算每一列 即跨行 因而axis=0
df.mean(axis = 0)
```
