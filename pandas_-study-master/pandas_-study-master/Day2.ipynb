# 一、Pandas数据查询

```python
import pandas as pd

file_path = r"D:\JupyterProject\study-pandas\file\datas\weather_20230115134249.csv"
df = pd.read_csv(file_path)
df.set_index('日期', inplace = True)
print(df.head())
print(df.index)
```

## 使用值列表批量查询

```python
#得到单个值
print(df.loc['2015-01-01', '城市'])
#传入单个参数索引名进行查询
print(df.loc["2015-01-01"])
#传入单个参数, 再跟着一个列表 ：查询某一行的多列
print(df.loc["2015-01-01", ["城市", "观测站","累积雨量(mm)"]])
#传入行索引的列表和列索引的列表：查询指定行的多列
print(df.loc[["2015-01-01", "2015-01-02"], ["城市", "观测站","累积雨量(mm)"]])
```

## 使用数字区间进行查询
**注意**： 
- 区间为闭区间
- 可以是不存在的索引名

```python
print(df.loc['199-01-03' : '2024-01-08', '城市'])
```

```python
print(df.loc['2015-01-03', "城市" : "观测站"])
```

```python
print(df.loc['2015-01-03': "2022-01-01", "城市" : "观测站"])
```

## 使用条件表达式查询
DataFrame 的中括号 [] 中可以填入一个 Series 类型的数据，并且这个 Series 数据通常是布尔类型，用于布尔索引（Boolean indexing）。这种布尔索引的作用是根据布尔值来选择DataFrame中的行或列。

布尔索引在Pandas中非常常见和有用。当你用一个布尔类型的 Series 对象对DataFrame进行索引时，只有在对应位置的布尔值为True时，相应的行或列才会被选中。如果布尔值为False，则不会包含在结果中。

python
import pandas as pd

# 创建DataFrame
df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'],
                   'Math': [80, 90, 75],
                   'English': [85, 95, 70],
                   'Science': [90, 85, 80]})

# 创建布尔Series
bool_series = pd.Series([True, False, True])

# 使用布尔索引选择行
selected_rows = df[bool_series]

print(selected_rows)
在上面的示例中，bool_series 是一个布尔类型的Series对象。在 df[bool_series] 这个语句中，只有在 bool_series 中对应位置的布尔值为True的行才会被选中，所以结果中只包含了第一行和第三行。

```python
print(df['相对湿度(%)'] < 70) #相当于对每一行数据 返回一个Series，其索引为原DataFrame索引，对应数据为True or False
```

```python
print(df.loc[df['相对湿度(%)'] < 70, :])
```
